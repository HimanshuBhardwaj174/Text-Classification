```python
import os

# Define the root directory where your data is stored
root_dir = "C:\\Users\\himan\\OneDrive\\Documents\\Jupyter\\Classical ML\\Text Classification\\20_newsgroups"

# Initialize empty lists to store newsgroups and articles
newsgroups = []
articles = []

# Loop through each directory (each newsgroup)
for newsgroup_dir in os.listdir(root_dir):
    newsgroup_path = os.path.join(root_dir, newsgroup_dir) ##first_sec
    if os.path.isdir(newsgroup_path):
        newsgroups.append(newsgroup_dir)  ##Target name
        
        # Loop through each file (each article) in the newsgroup directory
        for article_file in os.listdir(newsgroup_path):
            article_path = os.path.join(newsgroup_path, article_file)
            if os.path.isfile(article_path):
                with open(article_path, 'r', encoding='utf-8', errors='ignore') as file:
                    content = file.read().split()
                    articles.append((newsgroup_dir, content))
                    
            break
            
    break

# Now you have newsgroups list and articles list containing tuples of (newsgroup, content)
# You can process this data further as per your requirements
```

```python
target = []
final=[]
for i in os.listdir(path):
    X= []
    target.append(i)
    new_path = os.path.join(path,i)
    for j in os.listdir(new_path):
        final_path = os.path.join(new_path,j)
        with open(final_path) as file:
            content = file.read()
            word = re.split(pattern=r'\W+',string=content)
            X.append(word)
        
        
    final.append(X)
    
    
```
